### Variabel

Variabel berfungsi untuk menyimpan data atau nilai yang dapat digunakan dan diubah dalam program. Variabel memungkinkan kita menyimpan hasil perhitungan atau informasi untuk digunakan kembali.

Secara umum, untuk menggunakan variabel, Kamu perlu memulai dengan memilih nama yang ingin digunakan. Nama variabel sebaiknya pendek dan deskriptif. Nama variabel juga harus memenuhi beberapa syarat:

* Tidak boleh ada spasi (misalnya, `test var` tidak diperbolehkan)
* Hanya boleh berisi huruf, angka, dan garis bawah (misalnya, `test_var!` tidak diperbolehkan)
* Harus dimulai dengan huruf atau garis bawah (misalnya, `1_var` tidak diperbolehkan)

Kemudian, untuk membuat variabel, Anda perlu menggunakan tanda = untuk menetapkan nilai yang ingin diberikan.

## Functions

#### Definisi Function

`Function` (Fungsi) adalah sekumpulan kode yang akan dijalankan apabila dipanggil. Fungsi dapat mengambil nol atau lebih input dan mengembalikan output yang sesuai. Dalam Python, kita biasanya mendefinisikan fungsi menggunakan `def:`. Contoh sederhana dari function adalah sebagai berikut:

````python
# Function tanpa input
def hello_world():
    print("Hello, World!")

hello_world()
# Function dengan input
def double(x):
    return x * 2
print(double(5))
````

Dalam function, kita mengenal namanya parameter. Parameter adalah nilai kosong yang nantinya akan diisi oleh sebuah nilai yang dinamakan `argument`. Berikut adalah contoh penggunaan dari parameter dalam function

````python
def least_difference(a, b, c):
    diff1 = abs(a - b)
    diff2 = abs(b - c)
    diff3 = abs(a - c)
    return min(diff1, diff2, diff3)
````

Fungsi dimulai dengan sebuah header yang diawali oleh kata kunci `def`. Blok kode yang ditambahkan (indentasi) setelah tanda `:` akan dijalankan ketika fungsi tersebut dipanggil.

````python
print(
    least_difference(1, 10, 100),
    least_difference(1, 10, 10),
    least_difference(5, 6, 7), # Python mengizinkan penggunaan koma tambahan di akhir daftar argumen
)
````

#### Penggunaan function didalam function

Function Python adalah kelas satu, yang berarti kita dapat menetapkannya ke variabel dan meneruskannya ke dalam fungsi seperti argumen lainnya:

````python
def mult_by_five(x):
    return 5 * x

def call(fn, arg):
    """Panggil fungsi pada argumen"""
    return fn(arg)

def squared_call(fn, arg):
    """Panggil fungsi pada hasil memanggil fungsi pada argumen"""
    return fn(fn(arg))
square_num = squared_call(mult_by_five, 3)
print(
    call(mult_by_five, 2),
    square_num,
    sep='\n', # '\n' adalah karakter baris baru - memulai baris baru
)
````

#### Lambda

`Lambda` adalah fungsi anonim (tanpa nama) di Python yang digunakan untuk membuat fungsi sederhana dalam satu baris. Biasanya digunakan ketika kita membutuhkan fungsi cepat tanpa harus mendefinisikannya secara lengkap dengan `def`.

Sintaks dasar dari lambda adalah sebagai berikut:
```
lambda arg1, arg2, ... : expression
```

Contoh dari lambda:

````python
# Fungsi lambda untuk menjumlahkan dua angka
add = lambda x, y: x + y
print(add(3, 5))  # Output: 8
````

Kamu dapat menetapkan `lambda` ke variabel, meskipun kebanyakan orang akan memberi tahu bahwa kamu harus menggunakan `def` sebagai gantinya:

````python
another_double = lambda x: 2 * x       # tidak disarankan
def another_double(x): return 2 * x    # sangat disarankan
````

##### Penggunaan umum lambda

Digunakan dengan `map()`

````python
numbers = [1, 2, 3, 4, 5]
even = list(filter(lambda x: x % 2 == 0, numbers))
print(even)  # Output: [2, 4]
````

Digunakan dengan `filter()`

````python
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, numbers))
print(squared)  # Output: [1, 4, 9, 16]
````

Digunakan dengan `sorted()`

````python
data = [(1, 'a'), (3, 'c'), (2, 'b')]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)  # Output: [(1, 'a'), (2, 'b'), (3, 'c')]
````

#### Default Argument

`Default Argument` adalah argumen yang diberikan apabila nilai dari suatu function yang memiliki parameter tidak diisi, `Default argument` akan memberikan nilai default pada function

````python
def sapa(who="Alif"):
    print("Halo,", who)

sapa()
sapa(who="Alka")
# Kamu tidak perlu menspesifikan function seperti diatas, cukup masukkan value nya saja itu sudah cukup
sapa("Samuel")
````

## Boolean, Conditions & Conditional

#### Definisi Boolean

Boolean adalah tipe data dalam Python yang hanya memiliki dua nilai: True atau False. Boolean biasanya digunakan untuk nilai yang biasanya muncul sebagai hasil dari suatu operasi logika.

````python
x = 10
y = 5
hasil = x > y # variabel hasil akan bernilai True

print(hasil)
````

#### Definisi Conditions

Conditions (kondisi) adalah kombinasi dari variabel, nilai, operator, dan fungsi yang menghasilkan nilai boolean (True atau False). conditions digunakan dalam percabangan atau pernyataan logika untuk menentukan alur eksekusi program.

Operator perbandingan yang digunakan dalam conditions dari perhitungan matematika yaitu:
* `=` (sama dengan),
* `!=` (tidak sama dengan),
* `<` (lebih kecil),
* `<=` (lebih kecil atau sama dengan),
* `>` (lebih besar),
* `>=` (lebih besar atau sama dengan)

Operator logika yang digunakan yaitu:
- `and` (kedua kondisi harus True)
- `or` (salah satu kondisi harus True)
- `not` (membalik nilai kondisi True menjadi False, dan sebaliknya)


````python
x = 10
y = 20
if x < y:  # Kondisi: apakah x lebih kecil dari y?
    print("x lebih kecil dari y")  # Ini akan dicetak karena kondisi bernilai True
````

#### Definisi Conditional

Conditional adalah struktur kode yang dijalankan berdsarkan kondisi tertentu. Dengan conditional, kita bisa menjalankan blok kode tertentu hanya jika suatu kondisi terpenuhi (True), dan bisa mengeksekusi blok kode yang berbeda jika kondisinya tidak terpenuhi (False).

contoh: `if`, `elif`, dan `else`


````python
# if statement
age = 18
if age >= 18:
    print("Kamu sudah dewasa")
````

````python
# if-else statement
age = 16
if age >= 18:
    print("Kamu sudah dewasa")
else:
    print("Kamu belum dewasa")
````

````python
# if-elif-else statement
score = 75
if score >= 90:
    print("Grade A")
elif score >= 75:
    print("Grade B")
else:
    print("Grade C")
````

Seperti yang sudah dijelaskan sebelumnya, list bersifat `mutable` sehingga kita bisa mengubah list dengan fleksibel

````python
fruits.insert(1, 'melon')    # Menyisipkan elemen 'melon' di indeks 1
print(fruits) # Output = ['apple', 'melon', 'orange', 'cherry', 'watermelon', 'guava', 'grape']
````

##### Menghapus Elemen di dalam List

````python
fruits.remove('melon')  # Menghapus elemen 'melon'
print(fruits) # Output = ['apple', 'orange', 'cherry', 'watermelon', 'guava', 'grape']
````

````python
fruits.pop(2) # Menghapus elemen di indeks 2
print(fruits) # Output: ['apple', 'orange', 'watermelon', 'guava', 'grape']
````

Slicing Elemen di List

````python
fruits = ['apple', 'banana', 'cherry', 'watermelon', 'guava']  # List dengan elemen string
````

````python
print(fruits[1:4]) # Output: ['banana', 'cherry', 'watermelon']
print(fruits[:3]) # Output: ['apple', 'banana', 'cherry']
print(fruits[:]) # Output: ['apple', 'banana', 'cherry', 'watermelon', 'guava']
print(fruits[::-1]) # Output: ['guava', 'watermelon', 'cherry', 'banana', 'apple']
print(fruits[::2]) # Output: ['apple', 'cherry', 'guava']
print(fruits[1::2]) # Output: ['banana', 'watermelon']
print(fruits[-3:-1]) # Output: ['cherry', 'watermelon']
````

## Loops & list comprehension

Loop (perulangan) adalah struktur yang memungkinkan eksekusi berulang suatu blok kode selama kondisi tertentu terpenuhi. Python memiliki dua jenis loop utama: for loop dan while loop.

List comprehension adalah cara singkat untuk membuat list baru berdasarkan list lain. Ini memungkinkan kamu membuat list dengan loop dan kondisi dalam satu baris kode, sehingga lebih efisien dan mudah dibaca.

Sintaks dari list comperehension yaitu:
`[value atau matematika for item in iterable if condition]`


````python
# For Loop
fruits = ['apple', 'banana', 'cherry', 'watermelon', 'guava']
for i in fruits:
    print(i)
# Output:
# apple
# banana
# cherry
# watermelon
# guava
````

````python
# While loop
count = 0
while count < 3:
    print("Count is:", count)
    count += 1
# Output:
# Count is: 0
# Count is: 1
# Count is: 2
````

````python
# List Comprehension
numbers = [1, 2, 3, 4, 5]
even_numbers = [x for x in numbers if x % 2 == 0]
print(even_numbers)  # Output: [2, 4]
````

## Strings & dictionary

#### Strings

String adalah urutan karakter yang digunakan untuk menyimpan dan memanipulasi teks. Dalam Python, string didefinisikan menggunakan tanda kutip tunggal (`'...'`) atau ganda (`"..."`). Biasanya digunakan untuk menyimpan nama, alamat, atau kalimat.

Ciri-ciri string:
- String bersifat immutable, artinya nilainya tidak bisa diubah setelah didefinisikan.
- String diakses menggunakan indeks di mana indeks dimulai dari 0 untuk karakter pertama.

##### Pembuatan String dan Index

````python
name = "John"
greeting = 'Hello, World!'

print(name[0])  # Output: J
print(greeting[7])  # Output: W
````

##### Penggabungan String

````python
full_name = "John" + " " + "Doe"
print(full_name)
print("John","Doe")  # Output: John Doe
````

````python
print(f"hello {full_name}")
print("hello {0}, nice to meet you mr {1}".format("John","Doe"))
````

##### String Slicing

````python
print(greeting[0:5])  # Output: Hello
print(greeting[7:])   # Output: World!
````

##### Metode String Umum

````python
print(name.lower())  # Output: john (Pengubahan huruf jadi huruf kecil)
print(name.upper())  # Output: JOHN (Pengubahan huruf jadi huruf besar)
print(greeting.replace('World', 'Python'))  # Output: Hello, Python! (Mengganti substring lama dengan yang baru)
````

#### Dictionary

Dictionary adalah tipe data di Python yang digunakan untuk menyimpan pasangan key-value (kunci-nilai). Setiap key harus unik, dan value bisa berupa tipe data apa saja. Dictionary bersifat mutable

Ciri-ciri dictionary:
- Didefinisikan menggunakan kurung kurawal {}.
- Setiap elemen di dalam dictionary berupa pasangan key: value.
- Key harus bersifat immutable (seperti string, integer, atau tuple), sedangkan value bisa berupa tipe data apa saja.

````python
# Contoh dictionary
person = {
    'name': 'Alice',
    'age': 25,
    'city': 'New York'
}

print(person['name'])  # Output: Alice
print(person['age'])   # Output: 25
````

##### Penambahan Elemen ke Dictionary

````python
person['age'] = 26  # Mengubah nilai 'age'
person['job'] = 'Engineer'  # Menambah key baru 'job'
print(person) # Output: {'name': 'Alice', 'age': 26, 'city': 'New York', 'job': 'Engineer'}
````

##### Penghapusan Elemen dari Dictionary

````python
del person['city']  # Menghapus key 'city'
person.pop('job')   # Menghapus key 'job'
print(person)  # Output: {'name': 'Alice', 'age': 26}
````

##### Metode Umum Dictionary

````python
print(person.keys())  # Output: dict_keys(['name', 'age']) - Mengambil semua key dalam dictionary
print(person.values())  # Output: dict_values(['Alice', 26]) - Mengambil semua nilai dalam dictionary
print(person.items())  # Output: dict_items([('name', 'Alice'), ('age', 26)]) - mengambil pasangan key-value sebagai tuple.
````

## MA Methods Basic With Pandas

Moving Average (MA) merupakan metode analisis data time series dengan menghitung rata-rata dari dari nilai-nilai sebelumnya dalam rentang waktu tertentu dalam sebuah dataset.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
df = df.set_index('Date')
df
````

````python
df = df.drop('month', axis=1)
df
````

````python
# Fungsi untuk plotting
def plot_moving_average(df, price_column, ma_column):
    plt.figure(figsize=(12, 6))
    plt.plot(df[price_column], label=price_column, marker='o', color='black', alpha=0.5)
    plt.plot(df[ma_column], label=ma_column, marker='o')

    plt.title('Moving Average')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid()
    plt.show()
````

### Simple Moving Average

Simple Moving Average (SMA) adalah metode moving average yang paling sederhana dan paling umum digunakan. SMA dihitung dengan merata-ratakan sejumlah data dalam periode waktu tertentu. Setiap data dalam periode tersebut diberi bobot yang sama, tanpa ada penekanan lebih pada data yang lebih baru atau lebih lama.

\

Screenshot 2024-10-02 130205.png

````python
# Menghitung Moving Average (MA) dengan periode 3
df['SMA_30'] = df['Price'].rolling(window=30).mean()
df.head(50)
````

````python
plot_moving_average(df, price_column='Price', ma_column='SMA_30')
````

### Centered Moving Average

CMA atau rata-rata bergerak terpusat adalah moving average yang dihitung dengan merata-ratakan sejumlah data di sekitar titik pusat. Berbeda dengan moving average biasa yang menggunakan data dari periode sebelumnya, CMA menggunakan data sebelum dan sesudah titik pusat untuk menghasilkan rata-rata.

Misalnya, untuk data berurutan dari 1 sampai 5, CMA untuk periode 3 bisa dihitung dengan merata-ratakan nilai-nilai yang berada di sekitar pusat (misalnya, titik 2, 3, dan 4). Dalam hal ini, nilai pada titik 3 dihitung sebagai rata-rata dari tiga nilai tersebut.

\

image.png

````python
df['CMA_30'] = df['Price'].rolling(window=30, center=True).mean()
````

````python
plot_moving_average(df, price_column='Price', ma_column='CMA_30')
````

### Exponential Moving Average

EMA adalah jenis moving average yang memberikan bobot lebih besar pada data yang lebih baru, sehingga lebih responsif terhadap perubahan harga atau data terbaru dibandingkan dengan simple moving average. Dalam EMA, nilai rata-rata dihitung dengan menggunakan faktor pembobotan eksponensial. Bobot ini menurun secara eksponensial untuk setiap data historis, yang berarti data terbaru memiliki pengaruh yang lebih besar dalam perhitungan dibandingkan data yang lebih lama.

\

image.png

````python
df['EMA_30'] = df['Price'].ewm(span=30, adjust=False).mean()
````

````python
plot_moving_average(df, price_column='Price', ma_column='EMA_30')
````

### Weighted Moving Average

WMA adalah jenis moving average di mana setiap data dalam periode yang dihitung diberi bobot berbeda. Semakin baru datanya, semakin besar bobot yang diberikan. WMA berguna untuk memberikan perhatian lebih pada data terbaru dibandingkan data lama. Berbeda dengan EMA, di WMA bobot diberikan secara linear, bukan eksponensial.

\

image.png

````python
weights = np.arange(1, 31)
df['WMA_30'] = df['Price'].rolling(window=30).apply(lambda x: np.dot(x, weights) / weights.sum(), raw=True)
````

````python
plot_moving_average(df, price_column='Price', ma_column='WMA_30')
````
